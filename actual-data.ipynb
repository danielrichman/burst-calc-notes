```python
import time
import csv
import urllib

ept_url = "http://habitat.habhub.org/habitat/_design/ept/_list/csv/payload_telemetry/flight_payload_time?" \
          "include_docs=true&startkey=[%220ac025994b03898acde1cc1e03d3f897%22,%22757f6fccac0c21ce31021876bcc3fc2e%22]&" \
          "endkey=[%220ac025994b03898acde1cc1e03d3f897%22,%22757f6fccac0c21ce31021876bcc3fc2e%22,[]]&" \
          "fields=time,altitude"

def reltime(x):
    y = time.strptime(x, "%H:%M:%S")
    return (y.tm_hour * 60 + y.tm_min) * 60 + y.tm_sec
    
    
nova_26 = urllib.request.urlopen(ept_url)
nova_26 = map(lambda x: x.decode("ascii"), nova_26)
nova_26 = iter(csv.reader(nova_26))
assert next(nova_26) == ['time', 'altitude']
nova_26 = [(reltime(t), int(alt)) for t, alt in nova_26]
```

```python
launch = burst = max_alt = None
for i, (_, a) in enumerate(nova_26):
    if launch is None and a > 300:
        launch = i
    if max_alt is None or a > max_alt:
        max_alt = a
        burst = i
```

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats

plt.figure(figsize=(15, 8))
x, y = np.transpose(nova_26[launch:burst])

plt.plot(x, y)
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
print(slope, intercept, r_value, p_value, std_err)

plt.plot(x, intercept + x * slope)
```

```python
d = 10
fig, ax1 = plt.subplots(figsize=(15, 8))
ax1.plot(x[d/2:-d/2], (y[d:] - y[:-d])/(x[d:]-x[:-d]))
ax2 = ax1.twinx()
ax2.plot(x[d/2:-d/2], y[d/2:-d/2])
```
